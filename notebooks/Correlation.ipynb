```python
%matplotlib inline
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
FILE = '../data/yield/РЎРІРѕРґРЅР°СЏ РІРµРіРµС‚Р°С†РёСЏ.xlsx'
SHEET = '2014'
```

```python
xls = pd.ExcelFile(FILE, on_demand=True)
xls.sheet_names
```

```python
df = pd.read_excel(FILE, sheetname=SHEET, header=1)
df.head()
```

```python
df.columns
```

```python
NDVI_COLUMNS = [col for col in df.columns if 'РЅРµРґРµР»СЏ' in col]
CULTURE_COLUMN = 'РљСѓР»СЊС‚СѓСЂР° 2014'
YIELD_COLUMN = 'РЈСЂРѕР¶Р°Р№РЅРѕСЃС‚СЊ, С‚./РіР°.'

print(NDVI_COLUMNS)
```

```python
df_cultures = df.groupby(df[CULTURE_COLUMN]).mean()
```

```python
df_cultures.head()
```

```python
df_cultures[NDVI_COLUMNS].T.plot()
```

```python
df.shape
```

```python
df_wheat = df[df[CULTURE_COLUMN] == 'РџС€РµРЅРёС†Р° РѕР·РёРјР°СЏ']
```

```python
wheat_ndvi = df_wheat[NDVI_COLUMNS].T.mean().T
wheat_yield = df_wheat[YIELD_COLUMN]

df_wheat = pd.concat([wheat_ndvi, wheat_yield], axis=1)
df_wheat.columns = ['NDVI', 'Yield']
df_wheat.head()
```

```python
df_wheat.info()
```

```python
df_wheat = df_wheat.dropna()
df_wheat.info()
```

```python
np.corrcoef(df_wheat.NDVI, df_wheat.Yield)
```

```python
stats.ttest_ind(df_wheat.NDVI, df_wheat.Yield)
```
