```python
import numpy as np
import pandas as pd
```

```python
df12 = pd.DataFrame.from_csv('../meteo_csv/2012.csv')
df13 = pd.DataFrame.from_csv('../meteo_csv/2013.csv')
df14 = pd.DataFrame.from_csv('../meteo_csv/2014.csv')
df15 = pd.DataFrame.from_csv('../meteo_csv/2015.csv')
df16 = pd.DataFrame.from_csv('../meteo_csv/2016.csv')
```

```python
frames = [df12,df13,df14,df15,df16]
```

```python
for item in frames:
    print(item.shape)
```

```python
df = pd.concat(frames)
```

```python
df.head()
```

```python
df.drop(df.columns[[0, 1,4,6,8,10,12,13,14,15,16]], axis=1, inplace=True)
```

```python
year_list = []
month_list = []
day_list = []
```

```python
for index, row in df.iterrows():
    year_list.append(row['yearmoda']//10000)
    month_list.append((row['yearmoda']//100)%100)
    day_list.append(row['yearmoda']%100)
```

```python
s1 = pd.Series(year_list, name='year')
s2 = pd.Series(month_list, name='month')
s3 = pd.Series(day_list, name='day')
```

```python
df['year'] = s1
df['month'] = s2
df['day'] = s3
```

```python
df.drop('yearmoda', axis=1, inplace=True)
```
